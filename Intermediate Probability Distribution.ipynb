### Probability Distribution in DS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
```

Bernoulli Distribution

```python
# p = probability of success
# size = numbere of experiements
# rvs = random variables 
bern = stats.bernoulli.rvs(size = 10000, p = 0.5)
sns.distplot(bern, kde = False)
plt.title("Bernoulli Distribution")
plt.xlabel("Bernoulli Distribution")
plt.ylabel("Frequency")
```

```python
# p = probability of success
# size = numbere of experiements
# rvs = random variables 
bern = stats.bernoulli.rvs(size = 10000, p = 0.7)
sns.distplot(bern, kde = False)
plt.title("Bernoulli Distribution")
plt.xlabel("Bernoulli Distribution")
plt.ylabel("Frequency")
```

### Uniform Distribution

a possibly unlimited number of outcomes are allowed and all the events hold the same probability to take place. As an example, imagine the roll of a fair dice. In this case, there are multiple possible events which each of them having the same probability to happen.

```python
# size = number of experiements
# loc = starting point
# scale = end point

uniform = stats.uniform.rvs(size = 100, loc = 0, scale = 6)
sns.distplot(uniform, kde = True)
```

Here is the distribution obtained when we roll a dice 100 times

### Fair Dice uniform Distribution

```python
prob = np.full((6),1/6)
ch = np.arange(1,7) # ch is choice
plt.bar(ch, prob)
plt.title("Fair Dice uniform Distribution")
plt.xlabel("Choices")
plt.ylabel("Pribability")
ax = plt.gca()
ax.set_ylim([0,1])
```

### Loaded dice with non uniform distribution

```python
sampl = np.random.uniform(low=0, high=0.9, size=(6,))
face = [1,2,3,4,5,6]
plt.bar(face, sampl)
axes = plt.gca()
axes.set_ylim([0,1])
```

### Binomial Distribution

The Binomial Distribution can instead be thought as the sum of outcomes of an event following a Bernoulli distribution. The Binomial Distribution is therefore used in binary outcome events and the probability of success and failure is the same in all the successive trials

```python
# p is the probability that one of the events is going to take place
binom = stats.binom.rvs(size = 10000, n=10,p=0.5,random_state = 0)
pd.Series(binom).plot(kind='hist',bins=50)
```

Given multiple trials, each of them is independent of each other (the outcome of one trial doesn't affect another one).
Each trial can lead to just two possible results (eg. winning or losing), which have probabilities p and (1 - p).

#### Change the probability value to see the distribution , as the trials is independent, we may see diff distribution

```python
for i in range(3,10,3):
    bi = stats.binom.rvs(size = 10000, n=10, p = 0.1*i, random_state = 42)
    pd.Series(bi).plot(kind='hist',bins=20, label="p = {:f}".format(0.1*i))
    plt.xlabel("Random Variable")
    plt.ylabel("Frequency")
    plt.title("Binomial distribution for varying p")
    plt.legend()
                      
```
