### Bernouli Distribution

This distribution has only two possible outcomes and a single trial.
A simple example can be a single toss of a biased/unbiased coin. In this example, the probability that the outcome might be heads can be considered equal to p and (1 - p) for tails

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
```

```python
probs =np.array([0.75,0.25])
trial = [0,1]

plt.bar(trial, probs)
plt.title("Bernouli Distribution on coin",fontsize = 12)
plt.xlabel('probability')
plt.ylabel('coin outcome')
axes = plt.gca()
axes.set_ylim([0,1])
```

### Uniform Distribution

It can be derived from the Bernouli Distribution. But here, a possibly unlimited number of outcomes are allowed and all the events hold the same probability to take place.(uniformly)

#### Example of rolling a dice

```python
p = np.full((6),1/6)
roll = [1,2,3,4,5,6]
plt.bar(roll, p)
plt.xlabel("Probability")
plt.ylabel("Fair Dice uniform distribution")
axes = plt.gca()
axes.set_ylim([0,1])
```

### Binomial Distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, either Success or Failure. 

A simple example of a Binomial Distribution in action can be the toss of a biased/unbiased coin repeated a certain amount of times.


```python
# pmf(random_variable, number_of_trials, probability)

for p in range(3,10,3): # (start , stop, step)
    x = np.arange(0,25)
    binom = stats.binom.pmf(x, 20, 0.1*p)
    plt.plot(x,binom, '-o',label='p = {:.2f}'.format(0.1*p))
    plt.xlabel("Random variable")
    plt.ylabel("Probability")
    plt.title("Binomial Distribution of varying p")
    plt.legend()
```

### Normal(Gaussian) Distribution

Normal distribution, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean. In graph form, normal distribution will appear as a bell curve.

In addition to this, the sum of small random variables also turns out to usually follow a normal distribution (Central Limit Theorem).

```python
n = np.arange(-50,50)
mean = 0
normal = stats.norm.pdf(n, mean, 10)
plt.plot(n, normal)
plt.xlabel("Distribution")
plt.ylabel("Probability")
plt.title("Normal Distribution")
```

- The curve is symmetric at the centre
- The area under the distribution curve is equal to 1 (all the probabilities must sum up to 1).

- 68% of the data lies within one standard deviation of the mean
- 95% within two standard deviations of the mean
- 99.7% within three standard deviations of the mean.
